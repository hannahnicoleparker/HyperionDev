```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

insurance_data = pd.read_csv("insurance.csv")
insurance_data.head()
# read file and display first few lines using .head()
```

```python
# select rows from the dataframe for plotting
age = insurance_data['age']
charges = insurance_data['charges']

# Create a scatter plot
plt.scatter(age, charges)
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.title('Age vs Insurance Charges')

plt.show()
```

```python
# Convert the 'age' and 'charges' data to NumPy arrays and reshape them
age = np.array(age).reshape(-1, 1)
charges = np.array(charges)

# Create a linear regression model and fit it to the data
model = LinearRegression()
model.fit(age, charges)

# Get the slope and intercept of the best-fit line
slope = model.coef_[0]
intercept = model.intercept_

# Generate predicted values using the model
predictions = model.predict(age)

# Create a scatter plot of the data points
plt.scatter(age, charges, label='Actual Data')

# Plot the best-fit line
plt.plot(age, predictions, color='red', label='Best-Fit Line')

# Label axis and title
plt.scatter(age, charges)
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.title('Age vs Insurance Charges')

# Show the plot
plt.show()
```
